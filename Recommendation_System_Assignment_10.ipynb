```python
book=pd.read_csv('book.csv',encoding = "ISO-8859-1")
book
```

```python
book_df=book.drop(['Unnamed: 0'], axis=1)
book_df
```

```python
books = book_df.rename({'User.ID':'UserID','Book.Title':"BookTitle",'Book.Rating':"BookRating"},axis=1)
books
```

```python
books[books.duplicated()]
```

```python
books[books.duplicated(subset=["UserID","BookTitle"])]
```

```python
bookclean = books.drop_duplicates(subset=["UserID","BookTitle"])
bookclean
```

```python
bookclean.sort_values('UserID')
```

```python
len(bookclean['UserID'].unique())
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
book_reader = bookclean.pivot(index='UserID',columns='BookTitle',values='BookRating').reset_index(drop= True)
book_reader
```

```python
book_reader.fillna(0, inplace=True)
book_reader
```

```python
reader_sim =1-pairwise_distances(book_reader.values,metric='cosine')
reader_sim
```

```python
# Storing result in Dataframe
reader_sim_df=pd.DataFrame(reader_sim)
```

```python
#Set the index and column names to user ids 
reader_sim_df.index=bookclean.UserID.unique()
reader_sim_df.columns=bookclean.UserID.unique()
```

```python
reader_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(reader_sim, 0)
reader_sim_df.iloc[0:5, 0:5]
```

```python
# Most similar users
reader_sim_df.idxmax(axis=1)[0:5]
```
